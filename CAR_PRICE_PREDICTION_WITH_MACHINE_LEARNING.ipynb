```python
"""
Now dropping all the columns that are non linear
1.symboling
2.carheight
3.stroke
4.compressionratio
5.peakrpm """
```

```python
dataset.drop(["symboling","carheight","stroke","compressionratio","peakrpm"],inplace = True,axis =1)
```

```python
dataset.head()
```

# Train Test Split

```python
from sklearn.model_selection import train_test_split
```

```python
X = dataset.drop("price",axis =1 )
```

```python
y=dataset["price"]
```

```python
X.head()
```

```python
y.head()
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
X_train.head(),X_train.shape
```

```python
X_test.head(),X_test.shape
```

```python
y_train.head(),y_test.head(),y_train.shape,y_test.shape
```

# Linear Regression

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
lr.coef_,lr.intercept_
```

```python
lr.score(X,y)
```

```python
"""The score of our model is 81.77%"""
```

```python
y_pred = lr.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```

```python
"""The r2 score of our model is 0.6"""
```
